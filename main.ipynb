```python
from bayes_opt import BayesianOptimization
import numpy as np
from scipy.stats import ks_2samp
from sklearn.metrics import r2_score

# Geração de dados fictícios

# Frequência de expostos por idade (0-126)
expostos = np.random.randint(100, 1000, size=127)

# Lista de tábuas de mortalidade de base
tabuas_base = [
    np.random.uniform(0.01, 0.1, size=127),  # Tábua 1
    np.random.uniform(0.01, 0.1, size=127),  # Tábua 2
    np.random.uniform(0.01, 0.1, size=127),  # Tábua 3
    # Adicione mais tábuas conforme necessário
]

def aplicar_agravamento(tabua, fator):
    return tabua * fator

def aplicar_delay(tabua, delay_anos):
    return np.roll(tabua, delay_anos) if delay_anos != 0 else tabua

def calcular_aderencia(expostos, tabua):
    esperados = expostos * tabua
    r2 = r2_score(expostos, esperados)
    ks_stat = ks_2samp(expostos, esperados).statistic
    return r2, ks_stat

def funcao_objetivo(tabua_index, fator_agravamento, delay_anos):
    # Selecionar a tábua de mortalidade correspondente
    tabua_base = tabuas_base[int(tabua_index)]
    
    # Aplicar agravamento e delay
    tabua_var = aplicar_agravamento(tabua_base, fator_agravamento)
    tabua_var = aplicar_delay(tabua_var, int(delay_anos))
    
    # Calcular aderência
    r2, ks_stat = calcular_aderencia(expostos, tabua_var)
    
    # Retornar métrica combinada
    return r2 - ks_stat

# Configuração da otimização bayesiana
pbounds = {
    'tabua_index': (0, len(tabuas_base) - 1),  # Índice da tábua de mortalidade
    'fator_agravamento': (0.8, 1.2),  # Intervalo para o agravamento
    'delay_anos': (-5, 5)  # Intervalo para o delay
}

# Função para executar a Otimização Bayesiana uma vez
def executar_otimizacao(random_state):
    otimizador = BayesianOptimization(
        f=funcao_objetivo,
        pbounds=pbounds,
        random_state=random_state,  # Para reprodutibilidade em cada execução
    )
    
    otimizador.maximize(
        init_points=10,  # Pontos iniciais aleatórios
        n_iter=50,  # Número de iterações de otimização
    )
    
    return otimizador.max  # Retorna o melhor resultado encontrado

# Executar a Otimização Bayesiana múltiplas vezes e agregar os resultados
resultados = []
n_repeticoes = 10

for i in range(n_repeticoes):
    resultado = executar_otimizacao(random_state=i)
    resultados.append(resultado)

# Agregar os resultados
melhores_targets = [r['target'] for r in resultados]
melhores_tabuas = [int(r['params']['tabua_index']) for r in resultados]
melhores_fatores = [r['params']['fator_agravamento'] for r in resultados]
melhores_delays = [r['params']['delay_anos'] for r in resultados]

media_target = np.mean(melhores_targets)
media_tabua = int(np.round(np.mean(melhores_tabuas)))  # Índice da melhor tábua
media_fator = np.mean(melhores_fatores)
media_delay = np.mean(melhores_delays)

print("Média dos melhores targets:", media_target)
print("Índice da tábua média mais aderente:", media_tabua)
print("Média dos melhores fatores de agravamento:", media_fator)
print("Média dos melhores delays:", media_delay)
```
